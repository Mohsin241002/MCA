1.	Write a python script to create a perceptron and demonstrate AND operation use two input neurons and one output neurons initialize the random weights for two input neurons between 0.2 to 0.8 and display the error function (3 M)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# =============================================================================
# PROGRAM-1: PERCEPTRON SYNTAX
# =============================================================================

# Q1 & Q2: AND/OR Perceptron
class Perceptron:
    def __init__(self, learning_rate=0.1, n_iterations=100):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.errors = []
    
    def fit(self, X, y, weight_range=(0.2, 0.8)):
        # Initialize random weights in specified range
        self.weights = np.random.uniform(weight_range[0], weight_range[1], X.shape[1])
        self.bias = 0
        
        for i in range(self.n_iterations):
            errors = 0
            for idx, x_i in enumerate(X):
                # Forward pass
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = np.where(linear_output >= 0, 1, 0)
                
                # Update weights
                update = self.learning_rate * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update
                errors += int(update != 0.0)
            
            self.errors.append(errors)
            if errors == 0: break

# Training data
X_and = np.array([[0,0], [0,1], [1,0], [1,1]])
y_and = np.array([0, 0, 0, 1])
X_or = np.array([[0,0], [0,1], [1,0], [1,1]])
y_or = np.array([0, 1, 1, 1])

# Train AND (weights 0.2-0.8)
p_and = Perceptron()
p_and.fit(X_and, y_and, (0.2, 0.8))

# Train OR (weights 0.4-0.7)
p_or = Perceptron()
p_or.fit(X_or, y_or, (0.4, 0.7))

# Plot error function
plt.figure(figsize=(6,4))
plt.plot(p_and.errors, label='AND')
plt.plot(p_or.errors, label='OR')
plt.xlabel('Iterations')
plt.ylabel('Errors')
plt.legend()
plt.title('Perceptron Error per Epoch')
plt.tight_layout()

# Q3: NOT gate - Manual weights
# Weight = -1, Bias = 0.5
def not_gate(x):
    return 1 if (-1 * x + 0.5) >= 0 else 0

```
