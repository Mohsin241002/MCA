```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Configure matplotlib for larger, clearer plots
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["image.cmap"] = "gray"

# Absolute path to the image and a quick sanity check
image_path = "/Users/mohsin/MCA/Assignments/Computer Vision/images.jpg"
assert os.path.exists(image_path), f"Image not found at: {image_path}"

# Load as grayscale for Fourier analysis
img_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
assert img_gray is not None, "Failed to load image."

print(f"Image shape: {img_gray.shape}")
plt.imshow(img_gray, vmin=0, vmax=255)
plt.title("Original Image (Grayscale)")
plt.axis("off")
plt.show()
```

```python
# 1) Two-dimensional FFT and magnitude visualization (frequency domain)
f = np.fft.fft2(img_gray)
# Use log scaling for better visibility of small values
magnitude = np.log1p(np.abs(f))

plt.imshow(magnitude)
plt.title("FFT Magnitude (Unshifted)")
plt.axis("off")
plt.show()
```
