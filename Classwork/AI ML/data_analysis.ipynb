```python
import numpy as np
import seaborn as sns   
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import the dataset (assuming it's a CSV file named 'dataset.csv')
dataset = pd.read_csv('Karnataka-DistPop1.csv')
```

```python
plt.figure(figsize=(10, 6))
dataset.boxplot()
plt.title('Box Plot of Dataset')
plt.show()
```

```python
# Extract the independent variable (X) using iloc and colon
X = dataset.iloc[:, 0]  # This will select all rows and the first column ('District')
# Extract both 'District' and 'population' columns as a new DataFrame with two features
features = dataset[['District', 'population']]
print(features)
```

```python
# Find the number of unique values in each column of the dataset
unique_counts = dataset.nunique()
print(unique_counts)
```

```python
# Check for missing values in the dataset
print("Missing values count per column:")
print(dataset.isna().sum())

print("\nNot missing values count per column:")
print(dataset.notnull().sum())

print("\nMissing values (True indicates missing):")
print(dataset.isna())

print("\nNot missing values (True indicates present):")
print(dataset.notnull())  
```

```python
# Remove any rows with missing values from the dataset (if any)
cleaned_dataset = dataset.dropna()
print(cleaned_dataset)
```

```python
# Fill missing values in 'District' and 'population' columns with hard coded values

# For 'District', fill with a hard coded string (e.g., 'UnknownDistrict')
# For 'population', fill with a hard coded integer (e.g., 0)
filled_dataset = dataset.fillna({
    'District': 'UnknownDistrict',
    'population': 0
})

print(filled_dataset)
# Fill missing values using forward fill (ffill)
ffill_dataset = dataset.ffill()
print("\nDataset after forward fill (ffill):")
print(ffill_dataset)

# Fill missing values using backward fill (bfill)
bfill_dataset = dataset.bfill()
print("\nDataset after backward fill (bfill):")
print(bfill_dataset)
```

```python
# Replace NaN values with '00' in the dataset using replace
dataset_replaced = dataset.replace(np.nan, '00')
print(dataset_replaced)
```

```python
# Remove duplicate rows from the dataset, keeping the first occurrence
deduped_dataset = dataset.drop_duplicates(keep='first')
print(deduped_dataset)

# Remove rows that do not have at least 2 non-NA values (threshold=2)
deduped_thresh_dataset = dataset.drop_duplicates(keep='first').dropna(thresh=2)
print(deduped_thresh_dataset)
```

```python
# Encode the 'District' categorical column into category codes
dataset['District_cat'] = dataset['District'].astype('category').cat.codes
print(dataset[['District', 'District_cat']])
```

```python
from sklearn.model_selection import train_test_split

# Use 'District_cat' as feature and 'population' as target for splitting
X_feature = dataset[['District']]
y_target = dataset['population']

X_train, X_test, y_train, y_test = train_test_split(X_feature, y_target, test_size=0.2, random_state=42)

print("Training set size:", X_train.shape[0])
print("Test set size:", X_test.shape[0])
```
